```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch

# Set embedding dimension
d_k = 64  # Common embedding size in transformers

# Define hardcoded attention scores (manually chosen)
attn_scores = torch.tensor([
    [10, 20, 30, 40, 50],
    [15, 25, 35, 45, 55],
    [20, 30, 40, 50, 60],
    [25, 35, 45, 55, 65],
    [30, 40, 50, 60, 70]
], dtype=torch.float32)

# Compute softmax without scaling
softmax_no_scaling = torch.softmax(attn_scores, dim=-1)

# Compute scaled attention scores
scaled_attn_scores = attn_scores / np.sqrt(d_k)

# Compute softmax with scaling
softmax_with_scaling = torch.softmax(scaled_attn_scores, dim=-1)

# Plot heatmaps for comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.heatmap(attn_scores.numpy(), cmap="Blues", annot=True, fmt=".1f", ax=axes[0])
axes[0].set_title("Raw Attention Scores")

sns.heatmap(softmax_no_scaling.numpy(), cmap="Reds", annot=True, fmt=".2f", ax=axes[1])
axes[1].set_title("Softmax Without Scaling")

sns.heatmap(softmax_with_scaling.numpy(), cmap="Greens", annot=True, fmt=".2f", ax=axes[2])
axes[2].set_title("Softmax With Scaling")

plt.show()
```
